```python
# mark the polygons!
#marker = cv2.imread('marker.png')

polygon_csv = pd.read_csv('rio_test.csv')

polygon_csv = polygon_csv[polygon_csv.ImageId==1]
```

```python
polygon_csv['xy'] = polygon_csv.apply(lambda row: [row['X'],row['Y']], axis=1)

pts = np.array(polygon_csv.groupby('BuildingId')['xy'].apply(list).tolist())
```

```python
out = a

for i in polygon_csv.BuildingId.unique():
    cv2.polylines(out,[np.array(pts[i-1])],True,(0,255,255))

tiff.imshow(out)
```

```python
cv2.imwrite('out.tif', out)
```
